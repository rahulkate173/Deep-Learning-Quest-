# Gradient Descent

#### The optimization algorithm used for the gradient descent algorithm used to adjust the weights in backpropogation is acheived by updating the weights and here learning rate is the parameter present having the initial and the final value which is high at the start and decrease as it reach to the global optimum

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense,BatchNormalization
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples=500,noise=0.20,random_state=42)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
plt.title("Moons dataset")
plt.scatter(X[:,0],X[:,1],c=y)
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

```python
  def train_model(batch_size:int):

    model = Sequential([
          Dense(10,activation='relu',input_dim=2),
          Dense(5,activation='relu'),
          Dense(1,activation='sigmoid')
      ])
    model.compile(optimizer='adam',loss='binary_crossentropy')
    history = model.fit(
      X_train,y_train,
      epochs = 10,
      batch_size = batch_size
    )
    return model
```

## batch Graddient Descent
#### Hero only one time weight is updated by calculating the cost function for all the values in the data by using vectorization (dot product)
eg: 500 rows in data and 10 epochs
#### > number of time weight update is 10 x 1 = 10

```python
model = train_model(X_train.shape[0])
```

## Stochasitc Gradient Descent
#### Here the traing data is shuffled and one value randomly is selected and the weight is updated from that value
Eg:
500 x 10 = 5000 times weight is updted

```python
model = train_model(1)
```

## Mini batch Gradient Descent
#### here the data is divided into the batches and the training is done according to it
eg: 500/5 = 100 here the batch size = 5 and at the each batch the model is trained

```python
model = train_model(5)
```
