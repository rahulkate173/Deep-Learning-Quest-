## Implementing Basis of tensorflow

```python
import tensorflow as tf
```

```python
## constant to create and tensorflow tensor
tf.constant(2)
```

By default the dtype of the tensorflow is 32 bit because it is far more than enough fro neural network weights

```python
### building and tensor of ndarray
tf.constant([1,2,3])
```

```python
### shape of tensor
tf.constant([1,2,3]).shape
```

```python
### similar operations as numpy
t = tf.constant([[1.,2.,3.],[4.,5.,6.]])
print("tensor: \n",t)
print("---"*10)
print("Square of tensor: \n",tf.square(t))
print(f"Scaling in tensorflow: \n{t+10}")
```

```python
t[:,:]
```

```python
t[:,-1] ## all row last columns
```

```python
t[0,0]
```

```python
## tranpose in tensor : it not only transpose but also create the tensor
x = tf.transpose(t)
x
```

```python
### example 
import numpy as np
n = np.array([[1.,2.,3.],[4.,5.,6.]])
x = n.T
print(x)
```
