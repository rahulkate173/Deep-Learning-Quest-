```python
## importing essential libraries
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

## Binary Crossentropy
#### So this los functions is used when we have two classes either(0/1)
#### While using this loss function we use "sigmoid" as activation function
#### Loss function
image.png

```python
## load libraries
from sklearn.datasets import make_classification
## for binary cross entropy
np.random.seed(42)

X, y = make_classification(n_samples=1000, n_features=10, n_informative=6, n_redundant=2, n_classes=2, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
## building the model
import matplotlib.pyplot as plt
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_split=0.2, verbose=2)
# ploting
plt.title("train loss vs train val loss")
plt.plot(history.history['loss'],label=['loss'])
plt.plot(history.history['val_loss'],label=['val_loss'])
plt.xlabel("number of epochs")
plt.ylabel("loss")
plt.legend()
plt.grid(True)
plt.show()
```

## categorical Cross Entropy
#### So this loss fuction or cost fuction is used when we have multiclass labels in the output feature
#### But when using this activation function should be "softmax"
### loss function
image.png

```python
## loading and transforming data
np.random.seed(42)

data = load_iris()
X = data.data
y = to_categorical(data.target)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(y_train.shape[1], activation='softmax')
])

model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_split=0.2, verbose=2)

## ploting the data
plt.title("train loss vs train val loss")
plt.plot(history.history['loss'],label=['loss'])
plt.plot(history.history['val_loss'],label=['val_loss'])
plt.xlabel("number of epochs")
plt.ylabel("loss")
plt.legend()
plt.grid(True)
plt.show()
```

#### How categorical Cross Entropy works?
#### here the output feature is encoded first using one hot encoding and then for each row loss is calculated and
#### and at the end it gives the probability of all classes in the range (0 - 1)

## 3. Sparse Categorical Cross Entropy
#### So this loss function we use when we have nulticlass in output label and it leads to more fast computation for more classes
#### Here the output is label not the probability in the range (0-1) like the categorical cross entropy gives
#### Here also activation function is 'softmax'
image.png


```python
## loading and transforming data
from sklearn.datasets import load_wine
data = load_wine()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=data.target, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(len(np.unique(y)), activation='softmax')
])

model.compile(optimizer=Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_split=0.2, verbose=2)

plt.title("Train Loss vs Validation Loss")
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel("Number of Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test loss: {loss:.4f} | Test accuracy: {accuracy:.4f}")
```
