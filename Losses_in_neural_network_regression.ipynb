## BOSTON HOUSE DATASET

```python
## including and getting dataset
from sklearn.model_selection import train_test_split
from keras.datasets import boston_housing
(X_train,y_train),(X_test,y_test) = boston_housing.load_data()
X_train, X_val,y_train,y_val = train_test_split(X_train,y_train)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### lets look and clean the data
np.unique(np.isnan(X_train))
np.unique(np.isnan(X_test))
np.unique(np.isnan(y_train))
np.unique(np.isnan(y_test))
```

```python
### lets build the neural network
from keras.models import Sequential
from keras.layers import Dense,Flatten
from keras.activations import relu
from keras.optimizers import SGD
from keras.callbacks import EarlyStopping
```

```python
reg_model = Sequential([
    Dense(64,activation=relu,input_shape=(X_train.shape[1],)),
    Dense(32,activation=relu),
    Dense(16,activation=relu),
    Dense(1,activation='linear')
])
```

```python
reg_model.summary()
```

```python
reg_model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
history = reg_model.fit(
    X_train,y_train,
    epochs = 100,
    batch_size = 32,
    validation_data=(X_val,y_val),
)
```

## MAE (Mean Absolute Error)
### Similar to mse but just thing is that here insted of squaring the residual we take the magnitude of it
### loss function : |yi - yi^ |
### cost function
image.png

```python
## compiling the model
reg_model.compile(optimizer='adam', loss='mae', metrics=['mae'])
## training the model
history = reg_model.fit(
    X_train,y_train,
    epochs = 10,
    batch_size = 32,
    validation_data=(X_val,y_val),
)
## plotting the loss
plt.title("train loss vs train val loss")
plt.plot(history.history['loss'],label=['loss'])
plt.plot(history.history['val_loss'],label=['val_loss'])
plt.xlabel("number of epochs")
plt.ylabel("loss")
plt.legend()
plt.grid(True)
plt.show()
```

They are not converging this is possible because of low epochs

## Huber loss
### so this loss function is unique because if u have outlier it will behave like mean squared error (mse) and if no outlier then behave like mean absolute error (mae)
### cost function
image.png

### here daba is hyperparameter

```python
## compiling the model
reg_model.compile(optimizer='adam', loss='huber', metrics=['mae'])
## training the model
history = reg_model.fit(
    X_train,y_train,
    epochs = 10,
    batch_size = 32,
    validation_data=(X_val,y_val),
)
## plotting the loss
plt.title("train loss vs train val loss")
plt.plot(history.history['loss'],label=['loss'])
plt.plot(history.history['val_loss'],label=['val_loss'])
plt.xlabel("number of epochs")
plt.ylabel("loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
### lets tune the daba/delta of the huber loss
from keras.losses import Huber
## compiling the model
reg_model.compile(optimizer='adam', loss=Huber(delta=0.1), metrics=['mae'])
## training the model
history = reg_model.fit(
    X_train,y_train,
    epochs = 10,
    batch_size = 32,
    validation_data=(X_val,y_val),
)
## plotting the loss
plt.title("train loss vs train val loss")
plt.plot(history.history['loss'],label=['loss'])
plt.plot(history.history['val_loss'],label=['val_loss'])
plt.xlabel("number of epochs")
plt.ylabel("loss")
plt.legend()
plt.grid(True)
plt.show()
```

```python
huber_list = [1,0.1,0.01,0.001]
history_list = []
for loss in huber_list:
  print(f"---"*40)
  print(f"delta : {loss}")
  reg_model.compile(optimizer='adam', loss=Huber(delta=loss), metrics=['mae'])
  ## training the model
  history = reg_model.fit(
      X_train,y_train,
      epochs = 50,
      batch_size = 32,
      validation_data=(X_val,y_val),
  )
  history_list.append(history)


```

```python
## plotting the loss
fig , ax = plt.subplots(nrows=1,ncols=4,figsize=(20,8),sharey=True)
## delta = 1
ax[0].set_title(f"train loss vs train val loss delta_{huber_list[0]}")
ax[0].plot(history_list[0].history['loss'],label=['loss'])
ax[0].plot(history_list[0].history['val_loss'],label=['val_loss'])
ax[0].set_xlabel("number of epochs")
ax[0].set_ylabel("loss")
ax[0].legend()
ax[0].grid(True)
## delta = 0.1
ax[1].set_title(f"train loss vs train val loss delta_{huber_list[1]}")
ax[1].plot(history_list[1].history['loss'],label=['loss'])
ax[1].plot(history_list[1].history['val_loss'],label=['val_loss'])
ax[1].set_xlabel("number of epochs")
ax[1].set_ylabel("loss")
ax[1].legend()
ax[1].grid(True)
## delta = 0.01
ax[2].set_title(f"train loss vs train val loss delta_{huber_list[2]}")
ax[2].plot(history_list[2].history['loss'],label=['loss'])
ax[2].plot(history_list[2].history['val_loss'],label=['val_loss'])
ax[2].set_xlabel("number of epochs")
ax[2].set_ylabel("loss")
ax[2].legend()
ax[2].grid(True)
## delta = 0.001
ax[3].set_title(f"train loss vs train val loss delta_{huber_list[3]}")
ax[3].plot(history_list[3].history['loss'],label=['loss'])
ax[3].plot(history_list[3].history['val_loss'],label=['val_loss'])
ax[3].set_xlabel("number of epochs")
ax[3].set_ylabel("loss")
ax[3].legend()
ax[3].grid(True)
plt.show()
```
