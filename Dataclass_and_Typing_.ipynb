## Dataclass
### it is used for storing like insted of building and manual class and using __init__ functions to store the data. Dataclass does all this things for us just to use the ***@dataclass*** decorator and use type hint so its basic use is to store the data

```python
from dataclasses import dataclass
```

```python
@dataclass
class ImageMetaData:
  image_name : str
  height : float
  width : float
```

```python
ImageMetaData("book.jpeg",24.0,23.0)
```

### It will automatically create the __init__ , __reper__ ,__eqr , etc functions

```python
## we can convert the dataclass to dictonary
from dataclasses import asdict

print(type(asdict(ImageMetaData("book.jpeg",24.0,23.0))))
```

```python
asdict(ImageMetaData("book.jpeg",24.0,23.0))
```

```python
## they can be immutable also
@dataclass(frozen=True) # to make them immutable
class Point:
    x: int
    y: int
```

## Typing
### The typing module helps specify types of variables, function arguments, and return values.

```python
## basic examples
def value(a:int,b:int) -> int:
  return a + b
```

```python
## using more complex example
from typing import List

numbers : List[int] = [1,2,3,4]
print(type(numbers))
```

```python
numbers : List[int] = [1.0,2.0,3.0,4.0]
print(type(numbers))
print(numbers)
```

```python
from typing import Dict

student : Dict[str,int] = {"rahul":1,"atharva":2}
print("[INFO] Student : ",student)
```

```python
from typing import Tuple

point: Tuple[int, int] = ((2,3),(2,4))
```

```python
## Optional
## IT means str or none i.e str|None
from typing import Optional

name: Optional[str] = None
```
